```python
import glob
import json
import numpy as np

"final_testset"

def load_stats(stats_dir):
    root_dir = os.path.join("/home/kevin/Documents/master-thesis/logs/", stats_dir)

    stats_files = sorted(glob.glob(os.path.join(root_dir, "*", "all_test_stats.json")))

    print(stats_files)
    
    val_freq = 5

    # print(stats_files)

    stats = {}

    for file in stats_files:
        with open(file, "r") as f:
            # print(json.load(f))
            stats.update(json.load(f))
            
    return stats

models = [
        "m109375_ddffnew_do_last",
        "m109419_ddffnew_rnn_fc_cclast",
        "m109417_ddffnew_rnn_fc_ccconv",
        "m109418_ddffnew_rnn_stack_cclast",
        "0_m109416_ddffnew_rnn_stack_ccconv_testset",
        "0_m109416_ddffnew_rnn_stack_ccconv_testset",
        "m111494_ddffnew_rnn_stack_ccconv_cocdepth",
        "m111495_ddffnew_rnn_stack_ccconv_big",
        "m111497_ddffnew_rnn_stack_ccconv_big_cocdepth",
        "m103067_recurae_depth_4",
        "m104703_recurae_depth_4_cocdepth",
        "m105049_recurae_depth_4_big",
        "m105080_recurae_depth_4_big_cocdepth",
        "m110196_recurae_depth_numframetest_1",
        "m110195_recurae_depth_numframetest_2",
        "m110193_recurae_depth_numframetest_3",
        "m103067_recurae_depth_4",
        "m110194_recurae_depth_numframetest_5"
    ]

labels = [
    "DDFF",
    "LSTM FC Cat-Last",
    "LSTM FC Cat-Conv",
    "LSTM Stack Cat-Last",
    "LSTM Stack Cat-Conv",
    "Standard",
    "CoC-Dec",
    "Larger Data",
    "Larger Data \& CoC-Dec",
    "Standard",
    "CoC-Dec",
    "Larger Data",
    "Larger Data \& CoC-Dec",
    "1 Frame",
    "2 Frames",
    "3 Frames",
    "4 Frames",
    "5 Frames",
]

def get_model_stats_count(stats, model):
    return len(next(v for k,v in stats.items() if model in k))

def get_model_stats(stats, model, epoch, metric):
    val = next(v for k,v in stats.items() if model in k)[epoch][metric if metric != "rms" else "mse"]
    
    if metric == "rms":
        val = np.sqrt(val)
        
    return val

def get_loss_epoch(stats, model, metric, epoch):
    return get_model_stats(stats, model, epoch, metric)

def get_loss_avg(stats, model, metric):
    return np.mean([get_model_stats(stats, model, epoch, metric) for epoch in range(get_model_stats_count(stats, model))])

def float_to_str(metric, num):
    if "mse" in metric:
        s = f"\\num{{{num:.1e}}}"
    elif "rms" in metric:
        s = f"{num:.3f}"
    elif "acc" in metric:
        s = f"{num:.2f}"
    
    return "$" + s + "$"
    
```

```python
def print_stats(stats, models, labels):
    metrics = ["mse", "rms", "acc"]
    sep = " & "
    
    # lines = []
    
    for i in range(len(models)):
        line = [labels[i]]
        
        for metric in metrics:
            line.append(float_to_str(metric, get_loss_avg(stats, models[i], metric)))
        
        print(sep.join(line), r"\\")
```
